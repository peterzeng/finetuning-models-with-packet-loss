```python
from matplotlib import pyplot as plt
import pandas as pd
import json
```

```python
def plot_steps_vs_nodes(data, dataset_name, accuracy_threshold=0.75):
    fig, ax = plt.subplots(figsize=(8, 6), dpi=300)
    ax.set_title(f"Steps Required to Reach {accuracy_threshold*100}% Accuracy - {dataset_name}")
    ax.set_xlabel("Node Count")
    ax.set_ylabel("Steps to Accuracy")
    x = [2, 4, 6]
    ax.set_xticks(x)
    ax.set_xticklabels(["2", "4", "6"])
    ax.plot(x, data["0"], label="0%", marker='o', color='blue')
    ax.plot(x, data["0.001"], label="0.1%", marker='o', color='orange')
    ax.plot(x, data["0.005"], label="0.5%", marker='o', color='green')
    ax.plot(x, data["0.01"], label="1.0%", marker='o', color='red')
    ax.legend(title="Packet Loss Rate")
    ax.grid()
    plt.show()
```

```python
data = json.load(open("results/results.json"))
```

```python
def plot_steps_vs_loss_rate(data, dataset_name):
    fig, ax = plt.subplots(figsize=(8, 6), dpi=300)
    ax.set_title(f"Steps to Accuracy When Trained on 4 nodes- {dataset_name}")
    ax.set_xlabel("Packet Loss Rate (%)")
    ax.set_ylabel("Steps")
    x = [0.0, 0.1, 0.5, 1.0]
    ax.set_xticks(x)
    ax.set_xticklabels([str(i) for i in x])
    accuracies = list(data.keys())
    for i in range(len(accuracies)):
        acc = accuracies[i]
        ax.plot(x, data[acc], label=acc, marker='o')
    ax.legend(title="Accuracy")
    ax.grid()
    plt.show()
```

```python
data = {
    '60%':[250, 300, 300, 350],
    '70%':[500, 650, 900, 1200],
    '75%': [1600, 1600, 1750, 2200]
}
```

```python
plot_steps_vs_loss_rate(data, "MNLI")
```
